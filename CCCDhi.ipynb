```python
import easyocr #bai lam chuan
import pandas as pd
import re
import os

def clean_text(text):
    # Nếu phát hiện từ khóa kết thúc địa chỉ thì cắt luôn
    stop_keywords = ['Có giá trị', 'Giá trị', 'Valid until', 'Ngày', 'Date']
    for keyword in stop_keywords:
        if keyword in text:
            return text.split(keyword)[0].strip()
    # Nếu phát hiện ngày tháng thì cắt
    date_pattern = r'\d{1,2}/\d{1,2}/\d{2,4}'
    match = re.search(date_pattern, text)
    if match:
        return text[:match.start()].strip()
    return text.strip()

def get_clean_residence(lines):
    collecting = False
    address_lines = []

    for line in lines:
        line = line.strip()
        if re.search(r'(Nơi thường trú|Place of residence)', line, re.IGNORECASE):
            collecting = True
            continue  # Bỏ dòng tiêu đề

        if collecting:
            if any(keyword in line for keyword in ['Có giá trị', 'Date of expiry', 'Giá trị đến', 'Valid until']):
                break  # Ngừng lấy nếu gặp thông tin ngày hết hạn
            if line and not line.isdigit():  # Bỏ dòng chỉ có số
                address_lines.append(line)

    return ' '.join(address_lines).strip()

# Khởi tạo EasyOCR
reader = easyocr.Reader(['vi', 'en'])

# Đường dẫn ảnh
image_path = '//content/e0a6e49e-0b59-4da0-b455-3ca65afbbb66.jpg'

# OCR ảnh
results = reader.readtext(image_path, detail=0)

# Gom toàn bộ text
full_text = "\n".join(results)

print("=== FULL TEXT OCR ===")
print(full_text)
print("=====================")

def extract_info(text):
    info = {
        'Họ và tên/Full name': '',
        'Ngày sinh/Date of birth': '',
        'Giới tính/Sex': '',
        'Quốc tịch/Nationality': '',
        'Quê quán/Place of origin': '',
        'Nơi thường trú/Place of residence': '',
        'Số CCCD/No': '',
    }

    lines = [line.strip() for line in text.split("\n") if line.strip()]

    # Số CCCD
    for line in lines:
        match = re.search(r'\b\d{12}\b', line)
        if match:
            info['Số CCCD/No'] = match.group()
            break

    # Họ và tên
    for idx, line in enumerate(lines):
        if re.search(r'Họ và tên', line, re.IGNORECASE):
            if idx + 1 < len(lines) and len(lines[idx + 1].split()) >= 2:
                info['Họ và tên/Full name'] = lines[idx + 1]
            else:
                name = re.sub(r'Họ và tên\s*:?', '', line, flags=re.IGNORECASE).strip()
                info['Họ và tên/Full name'] = name
            break

    # Ngày sinh
    for line in lines:
        dob_match = re.search(r'\d{2}/\d{2}/\d{4}', line)
        if dob_match:
            info['Ngày sinh/Date of birth'] = dob_match.group()
            break

    # Giới tính
    if "Nam" in text:
        info['Giới tính/Sex'] = "Nam"
    elif "Nữ" in text:
        info['Giới tính/Sex'] = "Nữ"

    # Quốc tịch
    if "Việt Nam" in text:
        info['Quốc tịch/Nationality'] = "Việt Nam"

    # Quê quán
    for idx, line in enumerate(lines):
        if re.search(r'(Quê|Place of origin)', line, re.IGNORECASE):
            if idx + 1 < len(lines):
                next_line = lines[idx + 1].strip()
                if len(next_line) < 5 or next_line.isdigit():
                    if idx + 2 < len(lines):
                        next_line += " " + lines[idx + 2].strip()
                info['Quê quán/Place of origin'] = clean_text(next_line)
            break

   # Nơi thường trú
    for idx, line in enumerate(lines):
      if re.search(r'(Nơi thường trú|Place of residence)', line, re.IGNORECASE):
        residence_lines = []
        i = idx + 1
        stop_keywords = ['Có giá trị', 'Giá trị', 'Valid until', 'Ngày', 'Date', 'Date of expiry']
        while i < len(lines):
            current_line = lines[i].strip()
            lower_line = current_line.lower()
            # Nếu dòng chứa từ khóa dừng hoặc ngày tháng thì bỏ qua nhưng tiếp tục đọc tiếp
            if any(kw.lower() in lower_line for kw in stop_keywords) or re.search(r'\d{1,2}/\d{1,2}/\d{2,4}', current_line):
                i += 1
                continue
            # Nếu dòng ngắn quá hoặc chỉ là số thì bỏ qua
            if current_line.isdigit() or len(current_line) <= 2:
                i += 1
                continue
            # Nếu dòng có dấu , hoặc ít nhất 2 từ thì coi là địa chỉ
            if ',' in current_line or len(current_line.split()) >= 2:
                residence_lines.append(current_line)
            i += 1
        # Ghép địa chỉ lại
        residence_text = ' '.join(residence_lines)
        info['Nơi thường trú/Place of residence'] = clean_text(residence_text)
        break

    return info

# Trích xuất dữ liệu
info = extract_info(full_text)

# Tạo DataFrame
df = pd.DataFrame([info])

# Ghi thêm tên file gốc
df['Tên file'] = os.path.basename(image_path)

# Lưu ra Excel
output_file = 'thong_tin_cccd.xlsx'
df.to_excel(output_file, index=False)

print(f"✅ Đã lưu file {output_file}")

# Tự động tải nếu dùng Colab
try:
    from google.colab import files
    files.download(output_file)
except:
    pass
```

```python
import easyocr #OKE
import pandas as pd
import re
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Đường dẫn ảnh
image_path = '/content/e0a6e49e-0b59-4da0-b455-3ca65afbbb66.jpg'

# Khởi tạo EasyOCR
reader = easyocr.Reader(['vi', 'en'])

# Đọc ảnh gốc bằng OpenCV
image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Vẽ bounding boxes
results_detail = reader.readtext(image_path)
for (bbox, text, prob) in results_detail:
    pts = [tuple(map(int, point)) for point in bbox]
    cv2.polylines(image, [np.array(pts)], isClosed=True, color=(0, 255, 0), thickness=2)
    cv2.putText(image, text, pts[0], cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

# Hiển thị ảnh
plt.figure(figsize=(12, 12))
plt.imshow(image)
plt.axis('off')
plt.title('THẺ CCCD')
plt.show()

# ✅ Sửa lỗi OCR thường gặp
def correct_common_errors(text):
    corrections = {
        'do\n': 'Tự do\n',
        'Hanh phúc': 'Hạnh phúc',
        'Viêt Nam': 'Việt Nam',
        'Dale of expiry': 'Date of expiry',
        'Cógiá trị': 'Có giá trị',
    }
    for wrong, right in corrections.items():
        text = text.replace(wrong, right)
    return text

# ✅ Làm sạch địa chỉ
def clean_text(text):
    stop_keywords = ['Có giá trị', 'Giá trị', 'Valid until', 'Ngày', 'Date']
    for keyword in stop_keywords:
        if keyword in text:
            return text.split(keyword)[0].strip()
    date_pattern = r'\d{1,2}/\d{1,2}/\d{2,4}'
    match = re.search(date_pattern, text)
    if match:
        return text[:match.start()].strip()
    return text.strip()

# ✅ Trích xuất thông tin chính
# ✅ Trích xuất thông tin chính
def extract_info(text):
    info = {
        'Họ và tên/Full name': '',
        'Ngày sinh/Date of birth': '',
        'Giới tính/Sex': '',
        'Quốc tịch/Nationality': '',
        'Quê quán/Place of origin': '',
        'Nơi thường trú/Place of residence': '',
        'Số CCCD/No': '',
        'Ngày hết hạn/Date of expiry': ''
    }

    lines = [line.strip() for line in text.split('\n') if line.strip()]

    # Số CCCD
    for line in lines:
        match = re.search(r'\b\d{12}\b', line)
        if match:
            info['Số CCCD/No'] = match.group()
            break

    # Họ và tên
    for idx, line in enumerate(lines):
        if 'Họ và tên' in line:
            if idx + 2 < len(lines) and 'Full name' in lines[idx + 1]:
                info['Họ và tên/Full name'] = lines[idx + 2]
            elif idx + 1 < len(lines):
                info['Họ và tên/Full name'] = lines[idx + 1]
            break

    # Ngày sinh
    for line in lines:
        dob = re.search(r'\d{2}/\d{2}/\d{4}', line)
        if dob:
            info['Ngày sinh/Date of birth'] = dob.group()
            break

    # Giới tính
    if 'Nam' in text:
        info['Giới tính/Sex'] = 'Nam'
    elif 'Nữ' in text:
        info['Giới tính/Sex'] = 'Nữ'

    # Quốc tịch
    if 'Việt Nam' in text or 'Viet Nam' in text:
        info['Quốc tịch/Nationality'] = 'Việt Nam'

    # Quê quán
    place_origin = []
    in_origin = False
    for line in lines:
        if 'Quê quán' in line:
            in_origin = True
            continue
        if 'Nơi thường trú' in line:
            break
        if in_origin:
            place_origin.append(line)
    info['Quê quán/Place of origin'] = ', '.join(place_origin)

    # ✅ Nơi thường trú (Cải tiến)
    residence_lines = []
    start_idx = -1

    for i, line in enumerate(lines):
        if re.search(r'Nơi thường trú|Place of residence', line, re.IGNORECASE):
            start_idx = i
            break

    if start_idx != -1:
        for j in range(start_idx + 1, min(len(lines), start_idx + 10)):
            clean_line = lines[j].strip()
            if clean_line in ["Có", "Tự", '"', "giá", ""]:
                continue
            # Ưu tiên dòng chứa địa danh, kể cả số Ấp
            if re.search(r'\b(Ấp\s*\d*|xã|thôn|tổ|ấp|Lộc|Bình|Ninh|Phước|Thiện|Đồng|Hòa)\b', clean_line, re.IGNORECASE):
                residence_lines.append(clean_line)

        full_residence = ', '.join(residence_lines)
        full_residence = re.sub(r'\s{2,}', ' ', full_residence).strip(' ,')
        info['Nơi thường trú/Place of residence'] = full_residence

    # Ngày hết hạn
    for i, line in enumerate(lines):
        if any(x in line for x in ['Có giá trị đến', 'Date of expiry', 'Valid until', 'trị đến']):
            date_match = re.search(r'\d{2}/\d{2}/\d{4}', line)
            if date_match:
                info['Ngày hết hạn/Date of expiry'] = date_match.group()
            elif i + 1 < len(lines):
                date_match = re.search(r'\d{2}/\d{2}/\d{4}', lines[i + 1])
                if date_match:
                    info['Ngày hết hạn/Date of expiry'] = date_match.group()
            break

    return info

# === OCR ảnh và xử lý ===
results = reader.readtext(image_path, detail=0)
full_text = "\n".join(results)
full_text = correct_common_errors(full_text)

# In kiểm tra
print("=== THÔNG TIN ĐƯỢC TRÍCH RA ===")
print(full_text)
print("==================================")

# Trích xuất
info = extract_info(full_text)

# Tạo DataFrame và lưu
df = pd.DataFrame([info])
df['Tên file'] = os.path.basename(image_path)
output_file = 'thong_tin_cccd.xlsx'
df.to_excel(output_file, index=False)
print(f"✅ Đã lưu file {output_file}")

print("\n===== THÔNG TIN ĐỊNH DẠNG CCCD =====\n")
print("CỘNG HÒA XÃ HỘI CHỦ NGHĨA VIỆT NAM")
print("Độc lập - Tự do - Hạnh phúc\n")
print("SOCIALIST REPUBLIC OF VIETNAM")
print("Independence - Freedom - Happiness\n")
print("CĂN CƯỚC CÔNG DÂN")
print("Citizen Identity Card\n")

print(f"Số / No.:               {info['Số CCCD/No']}\n")

print("Họ và tên / Full name:")
print(f"{info['Họ và tên/Full name']}\n")

print("Ngày sinh / Date of birth:       Giới tính / Sex:       Quốc tịch / Nationality:")
print(f"{info['Ngày sinh/Date of birth']:<32}{info['Giới tính/Sex']:<22}{info['Quốc tịch/Nationality']}\n")

print("Quê quán / Place of origin:")
print(f"{info['Quê quán/Place of origin']}\n")

print("Nơi thường trú / Place of residence:")
print(f"{info['Nơi thường trú/Place of residence']}\n")

# Move this here
print("Ngày hết hạn / Date of expiry:")
print(f"{info['Ngày hết hạn/Date of expiry']}\n")
```
