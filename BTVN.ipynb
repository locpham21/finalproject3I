```python
#Bai2.2
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
sugarin=ctrl.Antecedent(np.arange(80,251.1,0.1),'sugarin')
BMI=ctrl.Antecedent(np.arange(18.5,30.1,0.1),'BMI')
WC=ctrl.Antecedent(np.arange(0,3,1),'WC')
fatigue=ctrl.Antecedent(np.arange(0,3,1),'fatigue')
diabetes=ctrl.Consequent(np.arange(0,4.1,0.1),'diabetes')
sugarin['low']=fuzz.trimf(sugarin.universe,[0,40,80])
sugarin['medium']=fuzz.trimf(sugarin.universe,[80,100,120])
sugarin['high']=fuzz.trimf(sugarin.universe,[120,160,200])
sugarin['very high']=fuzz.trimf(sugarin.universe,[160,250,250])
BMI['low']=fuzz.trimf(BMI.universe,[0,15,18.5])
BMI['medium']=fuzz.trimf(BMI.universe,[18.5,23,25])
BMI['high']=fuzz.trimf(BMI.universe,[25,29.9,29.9])
BMI['very high']=fuzz.trimf(BMI.universe,[29.9,30,30])

WC['low']=fuzz.trimf(WC.universe,[0,0,1])
WC['medium']=fuzz.trimf(WC.universe,[0,1,2])
WC['high']=fuzz.trimf(WC.universe,[1,2,2])
fatigue['low']=fuzz.trimf(fatigue.universe,[0,0,1])
fatigue['medium']=fuzz.trimf(fatigue.universe,[0,1,2])
fatigue['high']=fuzz.trimf(fatigue.universe,[1,2,2])
diabetes['LR']=fuzz.trimf(diabetes.universe,[0,0,1])
diabetes['MR']=fuzz.trimf(diabetes.universe,[0,1,2])
diabetes['HR']=fuzz.trimf(diabetes.universe,[2,2,3])
diabetes['VHR']=fuzz.trimf(diabetes.universe,[3,4,4])
rule11=ctrl.Rule(sugarin['very high']& WC['high'],diabetes['VHR'])
rule12=ctrl.Rule(sugarin['very high']& WC['medium'],diabetes['HR'])
rule13=ctrl.Rule(sugarin['very high']& WC['high'],diabetes['HR'])
rule14=ctrl.Rule(sugarin['medium']&WC['low'],diabetes['LR'])

rule21=ctrl.Rule(sugarin['very high']&BMI['very high'],diabetes['VHR'])
rule22=ctrl.Rule(sugarin['very high']&BMI['high'],diabetes['HR'])
rule23=ctrl.Rule(sugarin['high']&BMI['medium'],diabetes['HR'])
rule24=ctrl.Rule(sugarin['high']&BMI['high'],diabetes['MR'])
rule25=ctrl.Rule(sugarin['medium']&BMI['medium'],diabetes['LR'])

rule31=ctrl.Rule(sugarin['very high']&fatigue['high'],diabetes['VHR'])
rule32=ctrl.Rule(sugarin['high']&fatigue['high'],diabetes['HR'])
rule33=ctrl.Rule(sugarin['medium']&fatigue['low'],diabetes['LR'])
rule34=ctrl.Rule(sugarin['medium']&fatigue['low'],diabetes['LR'])

control_system=ctrl.ControlSystem([rule11,rule12,rule13,rule14,rule21,rule22,rule23,rule24,rule25,rule31,rule32,rule33,rule34])
fuzzy_system=ctrl.ControlSystemSimulation(control_system)
fuzzy_system.input['sugarin']=180
fuzzy_system.input['BMI']=25
fuzzy_system.input['WC']=2
fuzzy_system.input['fatigue']=1

fuzzy_system.compute()
print("Muc do canh bao:",fuzzy_system.output['diabetes'])
diabetes.view(sim=fuzzy_system)
```

```python
#Bai2.5
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
temperature=ctrl.Antecedent(np.arange(0,41,1),'temperature')
humidity=ctrl.Antecedent(np.arange(0,101,1),'humidity')
people=ctrl.Antecedent(np.arange(0,3.1,0.1),'people')
fan_speed=ctrl.Consequent(np.arange(0,101,1),'fan_speedd')
cooling_level=ctrl.Consequent(np.arange(0,101,1),'cooling_level')

temperature['C']=fuzz.trimf(temperature.universe,[0,0,20])
temperature['CL']=fuzz.trimf(temperature.universe,[10,15,20])
temperature['CM']=fuzz.trimf(temperature.universe,[10,25,25])
temperature['W']=fuzz.trimf(temperature.universe,[25,30,35])
temperature['H']=fuzz.trimf(temperature.universe,[30,40,40])

humidity['low']=fuzz.trimf(humidity.universe,[0,0,50])
humidity['normal']=fuzz.trimf(humidity.universe,[30,50,70])
humidity['high']=fuzz.trimf(humidity.universe,[50,100,100])

people['F']=fuzz.trimf(people.universe,[0,0,1])
people['M']=fuzz.trimf(people.universe,[1,1,2])
people['MA']=fuzz.trimf(people.universe,[2,3,3])

fan_speed['low']=fuzz.trimf(fan_speed.universe,[0,0,50])
fan_speed['normal']=fuzz.trimf(fan_speed.universe,[30,50,70])
fan_speed['high']=fuzz.trimf(fan_speed.universe,[50,100,100])

cooling_level['low']=fuzz.trimf(cooling_level.universe,[0,0,50])
cooling_level['medium']=fuzz.trimf(cooling_level.universe,[0,50,100])
cooling_level['high']=fuzz.trimf(cooling_level.universe,[50,100,100])

rule1=ctrl.Rule(temperature['C'],fan_speed['low'])
rule2=ctrl.Rule(temperature['CL'],fan_speed['low'])
rule3=ctrl.Rule(temperature['CM'],fan_speed['normal'])
rule4=ctrl.Rule(temperature['W'],fan_speed['high'])
rule5=ctrl.Rule(temperature['H'],fan_speed['high'])

rule6=ctrl.Rule(temperature['C'],cooling_level['low'])
rule7=ctrl.Rule(temperature['CL'],cooling_level['low'])
rule8=ctrl.Rule(temperature['CM'],cooling_level['medium'])
rule9=ctrl.Rule(temperature['W'],cooling_level['high'])
rule10=ctrl.Rule(temperature['H'],cooling_level['high'])

rule11=ctrl.Rule(humidity['low'],cooling_level['low'])
rule12=ctrl.Rule(humidity['normal'],cooling_level['medium'])
rule13=ctrl.Rule(humidity['high'],cooling_level['high'])

rule14=ctrl.Rule(people['F'],fan_speed['low'])
rule15=ctrl.Rule(people['M'],fan_speed['normal'])
rule16=ctrl.Rule(people['MA'],fan_speed['high'])

rule17=ctrl.Rule(temperature['H']&humidity['high']&people['MA'],cooling_level['high'])
rule18=ctrl.Rule(temperature['H']&humidity['high']&people['MA'],fan_speed['high'])

rule19=ctrl.Rule(temperature['CM']&humidity['normal']&people['M'],cooling_level['medium'])
rule20=ctrl.Rule(temperature['H']&humidity['normal']&people['MA'],fan_speed['normal'])

rule21=ctrl.Rule(temperature['C']&humidity['low']&people['F'],cooling_level['low'])
rule22=ctrl.Rule(temperature['C']&humidity['low']&people['F'],fan_speed['low'])

control_system=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22])
fan_speedd=ctrl.ControlSystemSimulation(control_system)

fan_speedd.input['temperature']=25
fan_speedd.input['humidity']=60
fan_speedd.input['people']=1

fan_speedd.compute()

temperature.view(sim=fan_speedd)
humidity.view(sim=fan_speedd)
fan_speed.view(sim=fan_speedd)
```

```python
#Grab3
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl


product_demand = ctrl.Antecedent(np.arange(0,11,0.1), 'product_demand')
competitor_pricing_pressure = ctrl.Antecedent(np.arange(0,11,0.1), 'competitor_pricing_pressure')
store_reputation = ctrl.Antecedent(np.arange(0,5.1,0.1), 'store_reputation')
profit_margin = ctrl.Antecedent(np.arange(0,11.1,0.1), 'profit_margin')
seasonal_demand = ctrl.Antecedent(np.arange(0,11.1,0.1), 'seasonal_demand')

discount = ctrl.Consequent(np.arange(0,70.1,0.1), 'discount')


product_demand['low'] = fuzz.trimf(product_demand.universe, [0,0,5])
product_demand['medium'] = fuzz.trimf(product_demand.universe, [0,5,10])
product_demand['high'] = fuzz.trimf(product_demand.universe, [5,10,10])

competitor_pricing_pressure['low'] = fuzz.trimf(competitor_pricing_pressure.universe, [0,0,5])
competitor_pricing_pressure['medium'] = fuzz.trimf(competitor_pricing_pressure.universe, [0,5,10])
competitor_pricing_pressure['high'] = fuzz.trimf(competitor_pricing_pressure.universe, [5,10,10])

store_reputation['low'] = fuzz.trimf(store_reputation.universe, [0,3.5,4])
store_reputation['medium'] = fuzz.trimf(store_reputation.universe, [3.5,4.2,4.5])
store_reputation['high'] = fuzz.trimf(store_reputation.universe, [4.2,4.5,5])

profit_margin['low'] = fuzz.trimf(profit_margin.universe, [0,0,5])
profit_margin['medium'] = fuzz.trimf(profit_margin.universe, [0,5,10])
profit_margin['high'] = fuzz.trimf(profit_margin.universe, [5,10,10])

seasonal_demand['none'] = fuzz.trimf(seasonal_demand.universe, [0,0,5])
seasonal_demand['medium'] = fuzz.trimf(seasonal_demand.universe, [0,5,10])
seasonal_demand['high'] = fuzz.trimf(seasonal_demand.universe, [5,10,10])


discount['very_low'] = fuzz.trimf(discount.universe, [0,0,5])
discount['low'] = fuzz.trimf(discount.universe, [0,5,10])
discount['medium'] = fuzz.trimf(discount.universe, [10,20,30])
discount['high'] = fuzz.trimf(discount.universe, [30,50,70])
discount['very_high'] = fuzz.trimf(discount.universe, [70,85,100])


rule1 = ctrl.Rule(product_demand['high'] & competitor_pricing_pressure['low'] & profit_margin['low'], discount['very_low'])
rule2 = ctrl.Rule(product_demand['low'] & competitor_pricing_pressure['high'] & profit_margin['high'], discount['high'])
rule3 = ctrl.Rule(store_reputation['high'] & profit_margin['medium'] & seasonal_demand['high'], discount['medium'])
rule4 = ctrl.Rule(competitor_pricing_pressure['high'] & seasonal_demand['high'] & profit_margin['high'], discount['very_high'])
rule5 = ctrl.Rule(store_reputation['low'] & product_demand['medium'] & profit_margin['low'], discount['medium'])
rule6 = ctrl.Rule(product_demand['high'] & seasonal_demand['none'] & competitor_pricing_pressure['low'], discount['very_low'])
rule7 = ctrl.Rule(profit_margin['high'] & competitor_pricing_pressure['medium'] & seasonal_demand['medium'], discount['medium'])


discount_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
discount_system = ctrl.ControlSystemSimulation(discount_ctrl)


discount_system.input['product_demand'] = 8
discount_system.input['competitor_pricing_pressure'] = 3
discount_system.input['store_reputation'] = 4.2
discount_system.input['profit_margin'] = 7
discount_system.input['seasonal_demand'] = 6


discount_system.compute()

discount.view(sim=discount_system)
print("Mức giảm giá:", discount_system.output['discount'])
```
