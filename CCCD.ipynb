```python
import easyocr
import pandas as pd
import re
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def clean_text(text):
    stop_keywords = ['C√≥ gi√° tr·ªã', 'Gi√° tr·ªã', 'Valid until', 'Ng√†y', 'Date']
    for keyword in stop_keywords:
        if keyword in text:
            return text.split(keyword)[0].strip()
    date_pattern = r'\d{1,2}/\d{1,2}/\d{2,4}'
    match = re.search(date_pattern, text)
    if match:
        return text[:match.start()].strip()
    return text.strip()

def get_clean_residence(lines):
    collecting = False
    address_lines = []

    for line in lines:
        line = line.strip()
        if re.search(r'(N∆°i th∆∞·ªùng tr√∫|Place of residence)', line, re.IGNORECASE):
            collecting = True
            continue

        if collecting:
            if any(keyword in line for keyword in ['C√≥ gi√° tr·ªã', 'Date of expiry', 'Gi√° tr·ªã ƒë·∫øn', 'Valid until']):
                break
            if line and not line.isdigit():
                address_lines.append(line)

    return ' '.join(address_lines).strip()

# Kh·ªüi t·∫°o EasyOCR
reader = easyocr.Reader(['vi', 'en'])

# ƒê∆∞·ªùng d·∫´n ·∫£nh
image_path = '//content/e0a6e49e-0b59-4da0-b455-3ca65afbbb66.jpg'

# üëá Hi·ªÉn th·ªã ·∫£nh CCCD tr∆∞·ªõc khi OCR
img = mpimg.imread(image_path)
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.title("·∫¢nh CCCD")
plt.show()

# OCR ·∫£nh
results = reader.readtext(image_path, detail=0)

# Gom to√†n b·ªô text
full_text = "\n".join(results)

print("=== FULL TEXT OCR ===")
print(full_text)
print("=====================")

def extract_info(text):
    info = {
        'H·ªç v√† t√™n/Full name': '',
        'Ng√†y sinh/Date of birth': '',
        'Gi·ªõi t√≠nh/Sex': '',
        'Qu·ªëc t·ªãch/Nationality': '',
        'Qu√™ qu√°n/Place of origin': '',
        'N∆°i th∆∞·ªùng tr√∫/Place of residence': '',
        'S·ªë CCCD/No': '',
    }

    lines = [line.strip() for line in text.split("\n") if line.strip()]

    for line in lines:
        match = re.search(r'\b\d{12}\b', line)
        if match:
            info['S·ªë CCCD/No'] = match.group()
            break

    for idx, line in enumerate(lines):
        if re.search(r'H·ªç v√† t√™n', line, re.IGNORECASE):
            if idx + 1 < len(lines) and len(lines[idx + 1].split()) >= 2:
                info['H·ªç v√† t√™n/Full name'] = lines[idx + 1]
            else:
                name = re.sub(r'H·ªç v√† t√™n\s*:?', '', line, flags=re.IGNORECASE).strip()
                info['H·ªç v√† t√™n/Full name'] = name
            break

    for line in lines:
        dob_match = re.search(r'\d{2}/\d{2}/\d{4}', line)
        if dob_match:
            info['Ng√†y sinh/Date of birth'] = dob_match.group()
            break

    if "Nam" in text:
        info['Gi·ªõi t√≠nh/Sex'] = "Nam"
    elif "N·ªØ" in text:
        info['Gi·ªõi t√≠nh/Sex'] = "N·ªØ"

    if "Vi·ªát Nam" in text:
        info['Qu·ªëc t·ªãch/Nationality'] = "Vi·ªát Nam"

    for idx, line in enumerate(lines):
        if re.search(r'(Qu√™|Place of origin)', line, re.IGNORECASE):
            if idx + 1 < len(lines):
                next_line = lines[idx + 1].strip()
                if len(next_line) < 5 or next_line.isdigit():
                    if idx + 2 < len(lines):
                        next_line += " " + lines[idx + 2].strip()
                info['Qu√™ qu√°n/Place of origin'] = clean_text(next_line)
            break

    for idx, line in enumerate(lines):
        if re.search(r'(N∆°i th∆∞·ªùng tr√∫|Place of residence)', line, re.IGNORECASE):
            residence_lines = []
            i = idx + 1
            stop_keywords = ['C√≥ gi√° tr·ªã', 'Gi√° tr·ªã', 'Valid until', 'Ng√†y', 'Date', 'Date of expiry']
            while i < len(lines):
                current_line = lines[i].strip()
                lower_line = current_line.lower()
                if any(kw.lower() in lower_line for kw in stop_keywords) or re.search(r'\d{1,2}/\d{1,2}/\d{2,4}', current_line):
                    i += 1
                    continue
                if current_line.isdigit() or len(current_line) <= 2:
                    i += 1
                    continue
                if ',' in current_line or len(current_line.split()) >= 2:
                    residence_lines.append(current_line)
                i += 1
            residence_text = ' '.join(residence_lines)
            info['N∆°i th∆∞·ªùng tr√∫/Place of residence'] = clean_text(residence_text)
            break

    return info

# Tr√≠ch xu·∫•t d·ªØ li·ªáu
info = extract_info(full_text)

# T·∫°o DataFrame
df = pd.DataFrame([info])
df['T√™n file'] = os.path.basename(image_path)

# L∆∞u ra Excel
output_file = 'thong_tin_cccd.xlsx'
df.to_excel(output_file, index=False)

print(f"‚úÖ ƒê√£ l∆∞u file {output_file}")

# T·ª± ƒë·ªông t·∫£i n·∫øu d√πng Colab
try:
    from google.colab import files
    files.download(output_file)
except:
    pass
```
