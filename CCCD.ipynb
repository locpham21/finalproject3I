```python
import easyocr
import pandas as pd
import re
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def clean_text(text):
    stop_keywords = ['Có giá trị', 'Giá trị', 'Valid until', 'Ngày', 'Date']
    for keyword in stop_keywords:
        if keyword in text:
            return text.split(keyword)[0].strip()
    date_pattern = r'\d{1,2}/\d{1,2}/\d{2,4}'
    match = re.search(date_pattern, text)
    if match:
        return text[:match.start()].strip()
    return text.strip()

def get_clean_residence(lines):
    collecting = False
    address_lines = []

    for line in lines:
        line = line.strip()
        if re.search(r'(Nơi thường trú|Place of residence)', line, re.IGNORECASE):
            collecting = True
            continue

        if collecting:
            if any(keyword in line for keyword in ['Có giá trị', 'Date of expiry', 'Giá trị đến', 'Valid until']):
                break
            if line and not line.isdigit():
                address_lines.append(line)

    return ' '.join(address_lines).strip()

# Khởi tạo EasyOCR
reader = easyocr.Reader(['vi', 'en'])

# Đường dẫn ảnh
image_path = '//content/e0a6e49e-0b59-4da0-b455-3ca65afbbb66.jpg'

# 👇 Hiển thị ảnh CCCD trước khi OCR
img = mpimg.imread(image_path)
plt.figure(figsize=(10, 10))
plt.imshow(img)
plt.axis('off')
plt.title("Ảnh CCCD")
plt.show()

# OCR ảnh
results = reader.readtext(image_path, detail=0)

# Gom toàn bộ text
full_text = "\n".join(results)

print("=== FULL TEXT OCR ===")
print(full_text)
print("=====================")

def extract_info(text):
    info = {
        'Họ và tên/Full name': '',
        'Ngày sinh/Date of birth': '',
        'Giới tính/Sex': '',
        'Quốc tịch/Nationality': '',
        'Quê quán/Place of origin': '',
        'Nơi thường trú/Place of residence': '',
        'Số CCCD/No': '',
    }

    lines = [line.strip() for line in text.split("\n") if line.strip()]

    for line in lines:
        match = re.search(r'\b\d{12}\b', line)
        if match:
            info['Số CCCD/No'] = match.group()
            break

    for idx, line in enumerate(lines):
        if re.search(r'Họ và tên', line, re.IGNORECASE):
            if idx + 1 < len(lines) and len(lines[idx + 1].split()) >= 2:
                info['Họ và tên/Full name'] = lines[idx + 1]
            else:
                name = re.sub(r'Họ và tên\s*:?', '', line, flags=re.IGNORECASE).strip()
                info['Họ và tên/Full name'] = name
            break

    for line in lines:
        dob_match = re.search(r'\d{2}/\d{2}/\d{4}', line)
        if dob_match:
            info['Ngày sinh/Date of birth'] = dob_match.group()
            break

    if "Nam" in text:
        info['Giới tính/Sex'] = "Nam"
    elif "Nữ" in text:
        info['Giới tính/Sex'] = "Nữ"

    if "Việt Nam" in text:
        info['Quốc tịch/Nationality'] = "Việt Nam"

    for idx, line in enumerate(lines):
        if re.search(r'(Quê|Place of origin)', line, re.IGNORECASE):
            if idx + 1 < len(lines):
                next_line = lines[idx + 1].strip()
                if len(next_line) < 5 or next_line.isdigit():
                    if idx + 2 < len(lines):
                        next_line += " " + lines[idx + 2].strip()
                info['Quê quán/Place of origin'] = clean_text(next_line)
            break

    for idx, line in enumerate(lines):
        if re.search(r'(Nơi thường trú|Place of residence)', line, re.IGNORECASE):
            residence_lines = []
            i = idx + 1
            stop_keywords = ['Có giá trị', 'Giá trị', 'Valid until', 'Ngày', 'Date', 'Date of expiry']
            while i < len(lines):
                current_line = lines[i].strip()
                lower_line = current_line.lower()
                if any(kw.lower() in lower_line for kw in stop_keywords) or re.search(r'\d{1,2}/\d{1,2}/\d{2,4}', current_line):
                    i += 1
                    continue
                if current_line.isdigit() or len(current_line) <= 2:
                    i += 1
                    continue
                if ',' in current_line or len(current_line.split()) >= 2:
                    residence_lines.append(current_line)
                i += 1
            residence_text = ' '.join(residence_lines)
            info['Nơi thường trú/Place of residence'] = clean_text(residence_text)
            break

    return info

# Trích xuất dữ liệu
info = extract_info(full_text)

# Tạo DataFrame
df = pd.DataFrame([info])
df['Tên file'] = os.path.basename(image_path)

# Lưu ra Excel
output_file = 'thong_tin_cccd.xlsx'
df.to_excel(output_file, index=False)

print(f"✅ Đã lưu file {output_file}")

# Tự động tải nếu dùng Colab
try:
    from google.colab import files
    files.download(output_file)
except:
    pass
```
